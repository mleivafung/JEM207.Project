```python
def path_to_image_html(path):
    """
    Convert the given path/link of the image to HTML tag.
    The return type of the function is the traditional IMG tag, that we use to render images on a webpage.
    """
    return '<img src="'+ path + '" width="60" >'
```

```python
def gross_number(table):
    """
    Transform a given vector from chr to float.
    The Input should be in the form of $XX,XXX,XXX.XX and the output would be in the form XXXXXXXX.X
    -------------------------------------------------------------------------------------------------
    Example:
    Input.-  [$24,765.9, $780,237,95] | Type: chr
    Output.- [24765.9, 78023795.0]   | Type: float
    """
    table = table
    
    i = len(table)
    while i > 0:
        table['Lifetime Gross'][i] = table['Lifetime Gross'][i][1:].replace(',','')
        i-=1
    
    i = len(table)
    while i > 0:
        table['Lifetime Gross'][i] = float(table['Lifetime Gross'][i])
        i-=1
        
    return table
```
